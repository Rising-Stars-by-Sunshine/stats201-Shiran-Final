Using the file `Explanation_Data.txt` generated by `PS2_Explanation_Query.ipynb` to generate a word cloud of all major keywords relating to data privacy/security in blockchain in the past 10 years.

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt

stopwords=set(STOPWORDS)
all_words=''
data_file=open('Explanation_Data.txt','r')
data=data_file.readlines()
data_file.close()
for line in data:all_words+=(line+'\n')

forbidden={'t','co','https'}
for word in forbidden:stopwords.add(word)

plt.figure(1,figsize=(12,8))
wordcloud=WordCloud(width=1200,height=800,background_color='white',stopwords=stopwords,min_font_size=10,collocation_threshold=float('inf')).generate(all_words)                
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
